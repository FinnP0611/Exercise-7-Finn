```python
oct1_temps = Helsinki_Vanta['TEMP'].loc[Helsinki_Vanta.index > '201910010000']
```

```python
ax2 = oct1_temps.plot(style='ro--', title='Helsinki-Vantaa Temps')
ax2.set_xlabel('Date')
ax2.set_ylabel('Temperature')
```

```python
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [12, 6]
```

```python
start_time = pd.to_datetime('201910011200')
end_time = pd.to_datetime('201910011500')
cold_time = pd.to_datetime('201910011220')

ax = oct1_temps.plot(style='ro--', title='Helsinki-Vantaa Temps', xlim=[start_time, end_time],
                     ylim=[40.0, 46.0])

ax.set_xlabel('Time')
ax.set_ylabel('Temperature (F)')

ax.text(cold_time, 42.0, '<- Coldest Time')
```
