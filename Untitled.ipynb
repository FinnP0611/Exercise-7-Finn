```python
# Filtering the Evening temps for oct 1st from 6 to 12pm into a new DF
oct1_evening_temps = Helsinki_Vanta['TEMP'].loc[Helsinki_Vanta.index >= '201910011800']

# Variable for the start, time end time, maximum temp in the timeframe
start_time = pd.to_datetime('201910011800')
end_time = pd.to_datetime('201910020000')
evening_max_index = oct1_evening_temps.idxmax()
evening_max = pd.to_datetime(evening_max_index)

```

```python
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [12, 6]

# start plotting
ax2 = oct1_evening_temps.plot(style='k:', 
                              title='Oct1 evening Temps Helsinki-Vantaa', 
                              xlim=[start_time, end_time]
                              )

ax2.set_xlabel('Datetime')
ax2.set_ylabel('Temperature (F)')
ax2.text(evening_max, oct1_evening_temps.max(), '<- warmest time')



```
